```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y_harvest, test_size=0.2, random_state=42
)
```

```python
harvest_model = RandomForestClassifier(
    n_estimators=100,
    max_depth=8,
    random_state=42
)

harvest_model.fit(X_train, y_train)
```

```python
farmer_input = pd.DataFrame([{
    "temperature_2m": 34,
    "relative_humidity_2m": 88,
    "rain": 0,
    "cloud_cover": 75,
    "wind_speed_10m": 3
}])
```

```python
harvest_pred = harvest_model.predict(farmer_input)[0]

if harvest_pred == 1:
    harvest_text = "Harvest in next 0–2 days"
else:
    harvest_text = "Harvest can wait 3–5 days"
```

```python
explainer = shap.TreeExplainer(harvest_model)
shap_values = explainer.shap_values(farmer_input)
```

```python
shap_df = pd.DataFrame({
    "Feature": farmer_input.columns,
    "Impact": shap_values[1][0]
}).sort_values(by="Impact", ascending=False)

shap_df
```
